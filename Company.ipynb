```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from openpyxl import load_workbook
from datetime import datetime, timedelta 
state='Bihar'
state_excel='Bihar.xlsx'
wb = load_workbook(state_excel)
ws = wb.worksheets[0]
driver = webdriver.Firefox(executable_path="geckodriver.exe")
driver.get('https://www.probe42.in/')
time.sleep(5)
name=driver.find_element_by_name('login')
name.send_keys('#ID')
password=driver.find_element_by_name('password')
password.send_keys('#password')
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[3]/input').click()
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[4]/input').click()
time.sleep(3)
driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div[4]/a').click()
driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[2]/div[1]/incdate').click()
state_input=driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[1]/div[1]/input')
state_input.send_keys(state)
date_init=datetime(2019,10,28)
date_final=datetime(2019,11,3)
date=date_init
driver.maximize_window()
while(date!=date_final):
    try:
        cin_list=[]
        backward_from='//*[@id="goto-search"]/div/div[2]/div[1]/div/div[1]/p/div/ul/li/div/div/div/table/thead/tr[1]/th[1]/button'
        forward_from='//*[@id="goto-search"]/div/div[2]/div[1]/div/div[1]/p/div/ul/li/div/div/div/table/thead/tr[1]/th[3]/button'
        driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[2]/div[1]/div/div[1]/p/dtcalendar/button/i').click()
        while(1):
            html1 = driver.page_source
            soup1=BeautifulSoup(html1,'html.parser')
            month_year=soup1.find_all(class_='btn btn-default btn-sm uib-title')[0].get_text()
            datetime_object = datetime.strptime(month_year, '%B %Y')
            if (month_year==date.strftime("%B %Y")):
                break
            elif (datetime_object>date):
                driver.find_elements_by_xpath(backward_from)[0].click()
                continue
            else:
                driver.find_element_by_xpath(forward_from)[0].click()
                continue
        if date.day<10:
            day=str('0'+str(date.day))
        else:
            day=str(date.day)
        id_req=[]
        html = driver.page_source
        soup=BeautifulSoup(html,'html.parser')
        for data in soup.find_all(class_='uib-day text-center'):
            if(data.text.strip()==day):
                id_req.append(data['id'])
        driver.find_element_by_id(id_req[0]).click() 
        current_date=date.strftime("%d %b %Y")
        box_date=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div[2]/form/div/div/div[2]/div[1]/div/div[1]/p/span').text.strip()
        if(box_date!=current_date):
            driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[2]/div[1]/div/div[1]/p/dtcalendar/button/i').click()
            box_date_formated=date.strptime(box_date,'%d %b %Y')
            if (date>box_date_formated):
                driver.find_element_by_xpath(forward_from).click()
                id_req=[]
                html = driver.page_source
                soup=BeautifulSoup(html,'html.parser')
                for data in soup.find_all(class_='uib-day text-center'):
                    if(data.text.strip()==day):
                        id_req.append(data['id'])
                driver.find_element_by_id(id_req[1]).click()        
            else:
                driver.find_element_by_xpath(backward_from)[0].click()
                id_req=[]
                html = driver.page_source
                soup=BeautifulSoup(html,'html.parser')
                for data in soup.find_all(class_='uib-day text-center'):
                    if(data.text.strip()==day):
                        id_req.append(data['id'])
                driver.find_element_by_id(id_req[1]).click()
        backward_to='/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div[2]/form/div/div/div[2]/div[1]/div/div[3]/p/div/ul/li/div/div/div/table/thead/tr[1]/th[1]/button'
        forward_to='/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div[2]/form/div/div/div[2]/div[1]/div/div[3]/p/div/ul/li/div/div/div/table/thead/tr[1]/th[3]/button'
        to='/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div[2]/form/div/div/div[2]/div[1]/div/div[3]/p/dtcalendar/button'
        driver.find_element_by_xpath(to).click()
        while(1):
            html1 = driver.page_source
            soup1=BeautifulSoup(html1,'html.parser')
            month_year=soup1.find_all(class_='btn btn-default btn-sm uib-title')[0].get_text()
            datetime_object = datetime.strptime(month_year, '%B %Y')
            if (month_year==date.strftime("%B %Y")):
                break
            elif (datetime_object>date):
                driver.find_elements_by_xpath(backward_to)[0].click()
                continue
            else:
                driver.find_element_by_xpath(forward_to)[0].click()
                continue
        if date.day<10:
            day=str('0'+str(date.day))
        else:
            day=str(date.day)
        id_req=[]
        html = driver.page_source
        soup=BeautifulSoup(html,'html.parser')
        for data in soup.find_all(class_='uib-day text-center'):
            if(data.text.strip()==day):
                id_req.append(data['id'])
        driver.find_element_by_id(id_req[0]).click() 
        current_date=date.strftime("%d %b %Y")
        box_date=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div[2]/form/div/div/div[2]/div[1]/div/div[3]/p/span').text.strip()
        if(box_date!=current_date):
            driver.find_element_by_xpath(to).click()
            box_date_formated=date.strptime(box_date,'%d %b %Y')
            if (date>box_date_formated):
                driver.find_element_by_xpath(forward_to).click()
                id_req=[]
                html = driver.page_source
                soup=BeautifulSoup(html,'html.parser')
                for data in soup.find_all(class_='uib-day text-center'):
                    if(data.text.strip()==day):
                        id_req.append(data['id'])
                driver.find_element_by_id(id_req[1]).click()        
            else:
                driver.find_element_by_xpath(backward_to)[0].click()
                id_req=[]
                html = driver.page_source
                soup=BeautifulSoup(html,'html.parser')
                for data in soup.find_all(class_='uib-day text-center'):
                    if(data.text.strip()==day):
                        id_req.append(data['id'])
                driver.find_element_by_id(id_req[1]).click() 
        search_button=driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[5]/div[4]/button[2]')
        search_button.click()
        time.sleep(2)
        while(1):
            try:
                driver.find_element_by_xpath('//*[@id="result-set"]/div/div[1]/div[1]/div/h5/span').click()
                time.sleep(4)
                continue
            except:
                break
        html1 = driver.page_source
        soup1=BeautifulSoup(html1,'html.parser')
        company_presence=len(soup1.find_all(class_='break-word'))
        if (company_presence==10):
            date=date+timedelta(days=1)
            continue
        else:
            company=int(soup1.find_all(class_='font-12-responsive')[-2].text)
            llp=int(soup1.find_all(class_='font-12-responsive')[-1].text)
            total=company+llp
            ten_company_pages=(total/10)
            rest=(total%10)
            if(ten_company_pages==1):
                html1 = driver.page_source
                soup1=BeautifulSoup(html1,'html.parser')
                for initial in range(10,38,3):
                    value=str(soup1.find_all(class_='break-word')[initial]).split('/')[4].split()[0][:-1]
                    transit_state=str(soup1.find_all(class_='break-word')[initial+1]).split()[-2]
                    if (transit_state==state):
                        cin_list.append(value)
                date=date+timedelta(days=1)
                for data in cin_list:
                    a=[data]
                    ws.append(a)
                wb.save(state_excel)    
                continue         
            else:
                allscan=int(ten_company_pages)
                if(allscan>0):
                    for i in range(0,allscan):
                        html1 = driver.page_source
                        soup1=BeautifulSoup(html1,'html.parser')
                        for initial in range(10,38,3):
                            value=str(soup1.find_all(class_='break-word')[initial]).split('/')[4].split()[0][:-1]
                            transit_state=str(soup1.find_all(class_='break-word')[initial+1]).split()[-2]
                            if (transit_state==state):
                                cin_list.append(value)
                        driver.find_element_by_link_text('›').click()
                    if (rest>0):
                        final=10+((rest-1)*3)
                        final=final+1
                        for i in range(10,final,3):
                            value=str(soup1.find_all(class_='break-word')[i]).split('/')[4].split()[0][:-1]
                            transit_state=str(soup1.find_all(class_='break-word')[i+1]).split()[-2]
                            if (transit_state==state):
                                cin_list.append(value)
                    date=date+timedelta(days=1)
                    for data in cin_list:
                        a=[data]
                        ws.append(a)
                    wb.save(state_excel)    
                    continue     
                
                else:
                    final=10+((rest-1)*3)
                    final=final+1
                    for i in range(10,final,3):
                        value=str(soup1.find_all(class_='break-word')[i]).split('/')[4].split()[0][:-1]
                        transit_state=str(soup1.find_all(class_='break-word')[i+1]).split()[-2]
                        if (transit_state==state):
                            cin_list.append(value)
                    
                    date=date+timedelta(days=1)
                    for data in cin_list:
                        a=[data]
                        ws.append(a)
                    wb.save(state_excel)
                    continue
    except:
        driver.close()
        driver = webdriver.Firefox(executable_path="geckodriver.exe")
        driver.get('https://www.probe42.in/')
        time.sleep(5)
        name=driver.find_element_by_name('login')
        name.send_keys('#ID')
        password=driver.find_element_by_name('password')
        password.send_keys('#Password')
        driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[3]/input').click()
        driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[4]/input').click()
        time.sleep(3)
        driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div[4]/a').click()
        driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[2]/div[1]/incdate').click()
        state_input=driver.find_element_by_xpath('//*[@id="goto-search"]/div/div[1]/div[1]/input')
        state_input.send_keys(state)
        driver.maximize_window()
        continue
print('Its done')
driver.close()
```
