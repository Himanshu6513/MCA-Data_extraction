```python
from bs4 import BeautifulSoup
from selenium import webdriver
import os
import time
from openpyxl import load_workbook
import xlsxwriter
import xlrd
import openpyxl
file_name='director_detail.xlsx' #Director Detail file
read_file='state.xlsx'           # State file to be read
required_list=[]
pan_detail=[]
din_detail=[]
driver = webdriver.Firefox(executable_path="geckodriver.exe")
driver.get('https://www.probe42.in/')
time.sleep(5)
name=driver.find_element_by_name('login')
name.send_keys('Username')
password=driver.find_element_by_name('password')
password.send_keys('Password')
driver.maximize_window()
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[3]/input').click()
driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[4]/input').click()
time.sleep(3)
driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ul/li[2]/a').click()
pan=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/span[3]')
din=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/span[2]')
pan_enter_key=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/inputbox[3]/div[1]/div/input')
din_enter_key=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/inputbox[2]/div[1]/div/input')
pan_search=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/inputbox[3]/div[1]/div/div[1]/button')
din_search=driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ui-view/div[1]/div/div/div/inputbox[2]/div[1]/div/div[1]/button')
time.sleep(2)
if os.path.exists(file_name):
    pass
else:
    outworkbook=xlsxwriter.Workbook(file_name)
    bold=outworkbook.add_format({'bold': True})
    detail=outworkbook.add_worksheet('Director_detail')
    features=['DIN','PAN','Name','DOB','Nationality','Gender','Address']
    column=0
    for feature in features:
        detail.write(0,column,feature,bold)
        column=column+1
    outworkbook.close()
wb = xlrd.open_workbook(read_file)      
sheet = wb.sheet_by_index(0)
if sheet.nrows>1:
    for i in range(1,sheet.nrows):
        required_list.append(sheet.cell_value(i,-1))
else:
    pass
sheet = wb.sheet_by_index(1)
if sheet.nrows>1:
    for i in range(1,sheet.nrows):
        required_list.append(sheet.cell_value(i,-1))
else:
    pass
sheet = wb.sheet_by_index(2)
if sheet.nrows>1:
    for i in range(1,sheet.nrows):
        required_list.append(sheet.cell_value(i,-1))
else:
    pass
wb = xlrd.open_workbook(file_name)      
sheet = wb.sheet_by_index(0)
if sheet.nrows>1:
    for i in range(1,sheet.nrows):
        din_detail.append(sheet.cell_value(i,0))
        pan_detail.append(sheet.cell_value(i,1))
else:
    pass
for item in required_list:
    if(item!='NA'):
        data=item.split(',')
    else:
        continue
    for value in data:
        while(1):
            try:
                update=[]
                if((value in pan_detail) or (value in din_detail)):
                    break
                else:
                    pass
                value_length=len(value)
                if(value_length==10):
                    pan.click()
                    pan_enter_key.send_keys(value)
                    pan_search.click()
                    time.sleep(2)
                    html = driver.page_source
                    soup=BeautifulSoup(html,'html.parser')
                    pan_enter_key.clear()
                    total_lengh=len(soup.find_all('td'))
                    if total_lengh>0:
                        update.append(soup.find_all('td')[0].text.split(':')[1].strip())
                        pan_detail.append(value)
                        update.append(soup.find_all('td')[1].text.split(':')[1].strip())
                        update.append(soup.find_all(class_='break-word')[0].text.strip())
                        update.append(soup.find_all('td')[2].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[3].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[4].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[5].text.split(':')[1].strip())
                        wb =openpyxl.load_workbook(file_name)
                        ws = wb.worksheets[0]
                        ws.append(update)
                        wb.save(file_name)
                        break
                            
                    else:
                        for i in range(0,7):
                            update.append('NA')
                        update[1]=value
                        pan_detail.append(value)
                        wb =openpyxl.load_workbook(file_name)
                        ws = wb.worksheets[0]
                        ws.append(update)
                        wb.save(file_name)
                        break
                else:
                    din.click()
                    din_enter_key.send_keys(value)
                    din_search.click()
                    time.sleep(2)
                    html = driver.page_source
                    soup=BeautifulSoup(html,'html.parser')
                    din_enter_key.clear()
                    total_lengh=len(soup.find_all('td'))
                    if total_lengh>0:
                        update.append(soup.find_all('td')[0].text.split(':')[1].strip())
                        din_detail.append(value)
                        update.append(soup.find_all('td')[1].text.split(':')[1].strip())
                        update.append(soup.find_all(class_='break-word')[0].text.strip())
                        update.append(soup.find_all('td')[2].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[3].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[4].text.split(':')[1].strip())
                        update.append(soup.find_all('td')[5].text.split(':')[1].strip())
                        wb =openpyxl.load_workbook(file_name)
                        ws = wb.worksheets[0]
                        ws.append(update)
                        wb.save(file_name)
                        break
                    else:
                        for i in range(0,7):
                            update.append('NA')
                        update[0]=value
                        din_detail.append(value)
                        wb =openpyxl.load_workbook(file_name)
                        ws = wb.worksheets[0]
                        ws.append(update)
                        wb.save(file_name)
                        break
                    
            except:
                driver.close()
                driver.get('https://www.probe42.in/')
                time.sleep(5)
                name=driver.find_element_by_name('login')
                name.send_keys('Username')
                password=driver.find_element_by_name('password')
                password.send_keys('Password')
                driver.maximize_window()
                driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[3]/input').click()
                driver.find_element_by_xpath('/html/body/div/div[2]/div[1]/div[1]/div[3]/form/div[4]/input').click()
                time.sleep(3)
                driver.find_element_by_xpath('/html/body/div/div[2]/div[2]/div/ui-view/div/ul/li[2]/a').click()
                continue
driver.close()
print('Its is done')
```
